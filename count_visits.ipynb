```python
from astropy import table
from newport import *
import numpy as np
```

```python
# binned - visits
# & HD 86226 & 967 & 1194 & 1159 & 1179 \\
for fn in target_fn:
    print(f"& {fn.replace('_', ' ')} ", end='')
    for band in ['B', 'V', 'R', 'I']:
        try:
            phot_table = table.Table.read(f'tables/binned/binned_{fn}_{band}.fits')
        except FileNotFoundError:
            continue
        # print(band, len(np.unique(phot_table['night'])))
        print(f"& {len(np.unique(phot_table['night']))} ", end='')
    print(' \\\\')
```

```python
# relphot - exposures
# & HD 86226 & 967 & 1194 & 1159 & 1179 \\
for fn in target_fn:
    print(f"& {fn.replace('_', ' ')} ", end='')
    for band in ['B', 'V', 'R', 'I']:
        try:
            phot_table = table.Table.read(f'tables/relphot/relphot_{fn}_{band}.fits')
        except FileNotFoundError:
            print("& 0 ", end='')
            continue
        # print(band, len(np.unique(phot_table['night'])))
        print(f"& {len(np.unique(phot_table['time']))} ", end='')
    print(' \\\\')
```
